##1. Bank Account Manager

```python
balance = 0.0   #global variable to store account balance
#function to deposit amount
def deposit(amount):
  global balance
  balance += amount
  print(f'{amount} deposited successfully.')
  print(f'Total Balance :{balance}\n')

#function to withdraw amount
def withdraw(amount):
  global balance
  if balance >= amount:
    balance -= amount
    print(f'{amount} withdraw successfully.')
  else:
    print('Insufficient Balance for Withdrawal!')
  print(f'Total Balance :{balance}\n')

#function to check balance
def Checkbalance():
  print(f'Your Current Balance is :{balance}\n')

  #-----main program---------
print('welcome to Smart Bank Account System')

while True:
  print('1 for deposit ')
  print('2 for withdraw ')
  print('3 for check balance ')
  print('4 for exit ')

  choice = input('Enter your choice (1-4): ')
  if choice == '1':
    amount = float(input('Enter amount to deposit:'))
    deposit(amount)
  elif choice == '2':
    amount = float(input('Enter amount to withdraw:'))
    withdraw(amount)
  elif choice =='3':
    Checkbalance()
  elif choice == '4':
    print('Thank you for using our service!')
    break
  else:
    print('invalid choice ... please Try again.\n')

```

##2. Weather Data Analyzer

```python
#function to calculate average temperature
def average_temp(temps):
  return sum(temps)/len(temps)

#function to calculate average humidity
def average_humidity(humidity):
  return sum(humidity)/len(humidity)

#function to determine weather conditions
def weather_condition(temp,humidities):
  if temp > 30 and humidities < 50:
    return 'Sunny'
  elif temp <20:
    return 'Cold'
  elif humidities >70:
    return "Rainy"
  else:
    return "Moderate"

#---------main program--------
temperatures = [30,34,29,30,31]    # in celsius
humidities = [72,77,60,70,80]     # in percentage

avg_temp = average_temp(temperatures)
avg_humidity = average_humidity(humidities)
condition = weather_condition(avg_temp,avg_humidity)

print("Weather Data Analyzer \n")
print(f'Average Temperature : {avg_temp:.2f}°C')
print(f'Average Humidity : {avg_humidity:.2f}%')
print(f'Overall Condition: {condition}')
```

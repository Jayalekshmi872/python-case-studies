#NumPy Problems

##Basic Array Operations

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('auto-mpg.csv')  #reading file
mpg = data['mpg'].to_numpy()   # convert mpg column into numpy array
```

```python
#basic statistics
mpg_mean = mpg.mean()  #mean of mpg
print('mean: ',mpg_mean)

mpg_median = np.median(mpg) #median of mpg
print('median: ',mpg_median)

mpg_std = mpg.std() #standard deviation of mpg
print('standard deviation :',mpg_std)
```

```python
#count of cars with mpg > 25

count_car = np.sum(mpg>25)
print('count of car with mpg >25 :', count_car)
```

##Filtering

```python
#convert to numpy arrays

cylinders = data['cylinders'].to_numpy()
car_names = data['car name'].to_numpy()
```

```python
#filter condition
filter_name = car_names[cylinders > 6]

 #convert to python list
filter_name = filter_name.tolist()
filter_name
```

##Statistical Analysis

```python
#convert to numpy
weight = data['weight'].to_numpy()

#calculate percentiles
p25 = np.percentile(weight,25)
p50 = np.percentile(weight,50)
p75 = np.percentile(weight,75)

print('25th percentile: ', p25)
print('50th percentile: ',p50)
print('75th percentile: ',p75)
```

##Array Manipulation

```python
#convert accelaration to numpy
acc = data['acceleration'].to_numpy()
#normalize to range 0-1
acc_norm = acc - acc.min()/acc.max() - acc.min()
acc_norm
```

##Broad Casting

```python
# Handle missing horsepower values (shown as '?')
horsepower_array = np.array(data['horsepower'])
# Convert '?' to NaN and handle missing values
horsepower_numeric = np.where(horsepower_array == '?', np.nan, horsepower_array)
horsepower_numeric = horsepower_numeric.astype(float)
mean_horsepower = np.nanmean(horsepower_numeric)
horsepower_filled = np.nan_to_num(horsepower_numeric, nan=mean_horsepower)
horsepower_increased = horsepower_filled * 1.10
print("Horsepower increased by 10% (first 10):", horsepower_increased[:10])
```

##Boolean Indexing

```python
#convert column to numpy
displacement = data['displacement'].to_numpy()
orgin =  data['origin'].to_numpy()
#boolean condition :orgin=2(Europe)
cond = orgin == 2
#filter displacement values from Europe cars
europe_displacement = displacement[cond]
#calulate average displacement
avg_displacement = displacement[cond].mean()
print('average displacement for European cars :',avg_displacement)
```

##Matrix Operations

```python
#covert required column to numpy array
matrix = data[['mpg','horsepower','weight']].to_numpy()
#given vector
vector =np.array([1,0.5,-0.2])
#dot product
result = np.dot(matrix,vector)
print(result)
```

##Sorting


```python
#convert to numpy
model_year = data['model year'].to_numpy()
car_names = data['car name'].to_numpy()
#sort by year decreasing
idx = np.argsort(-model_year)
#print first 5 cars
print(car_names[idx][:5])
```

##Correlation

```python
# convert to numpy array
mpg = data['mpg'].to_numpy()
weight = data['weight'].to_numpy()
#compute correlation matrix
corr_matrix = np.corrcoef(mpg,weight)[1,0]
print('correlation between mpg and weight :',corr_matrix )
```

##Conditional Aggregates

```python
#convert to numpy
cylinder = data['cylinders'].to_numpy()
#find unique cylender values
unique_cyl = np.unique(cylinder)
#loop and calculate mean mpg per cylender group
for c in unique_cyl:
  mean_mpg = mpg[cylinder == c].mean()
  print(f'cylenders: {c}, average mpg: {mean_mpg}')
```

##Basic Exploration

```python
import pandas as pd

#load dataset
df = pd.read_csv('auto-mpg.csv')
df
```

```python
# 2. total number of rows and columns

print( '\ntotal rows and columns:')
df.shape
```

##Filtering and Indexing

```python
#filter cars from 1975 and weight <3000
filtered_df = df[(df['model year']==1975) & (df['weight']<3000)]
#show selected columns only
result = filtered_df[['car name','weight','mpg']]
result
```

##Handling Missing Data

```python
#check for missing values in df
check_miss = df.isna().sum()
check_miss

#result :No missing value found ,so no replacement needed
```

##Data Transformation

```python
#convert string to numeric type
df['horsepower']= pd.to_numeric(df['horsepower'],errors='coerce')
```

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
df.head()
```

##Group By

```python
#group cars by orgin
group = df.groupby('origin')
#calculate mean mpg
mpg_mean = group['mpg'].mean()
mpg_mean
```

##Sorting

```python
#sort mpg column in decending order
sorted_df = df.sort_values(by = 'mpg',ascending=False)
#select top 10 cars after sorting
sorted_df.head(10)
```

##Apply Function

```python
# Define the custom function to calculate performance score
def performance_score(row):
    # Formula: mpg * acceleration / weight
    return row['mpg'] * row['acceleration'] / row['weight']

# Apply the function to each row and create a new column
df['performance_score'] = df.apply(performance_score, axis=1)

# Display first few rows to confirm
df.head()
```

##Visualization Preparation

```python
#group by model year and calculate average mpg, weight and horsepower
summary_df = df.groupby('model year')[['mpg','weight','horsepower']].mean()
#display the summary dataframe
summary_df
```

##Exporting Data

```python
#filter cars with mpg > 30
high_mpg_cars = df[df['mpg']>30]
#select only required columns
high_mpg_cars = high_mpg_cars[['mpg','cylinders','horsepower','weight']]
#save to csv file
high_mpg_cars.to_csv('high_mpg_cars.csv')
#display the filtered dataframe
high_mpg_cars
```

##Finding Anomalies

```python
# Calculate Q1 and Q3 for mpg
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)

# Calculate Interquartile Range (IQR)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Filter outliers in mpg
outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]

# Select required columns
outliers = outliers[['car name', 'mpg', 'model year']]

# Display outliers
outliers
```
